## Group By and Aggregating

```python
group_by_frame = df.groupby('Revenue growth')
group_by_frame.mean() # It accur an error because Revenue growth is object data type mean function work on numeric data type
```

```python
# .count(): This is an aggregation function that you apply to each of 
# those groups created in step 1. Specifically, count() calculates
# the number of non-missing (non-NaN) values within each group for each column 
# (excluding the grouping column itself, 'Name' in this case).
df.groupby('Name').count
```

```python
#We use .min() to find the minimum value within each group. 
# When your data is grouped by a column (like 'Name'), .min()
# helps you get the smallest value in each group for the other columns.
# Same as max
df.groupby('Name').min()
```

```python
df.groupby('Name').sum()
```

```python
#Data Exploration: 
#It provides a quick overview of the distribution and spread of data within each group.
#Detecting Outliers: 
#By looking at the min and max values, you can spot outliers in each group
#Summary Statistics:
#Itâ€™s a great way to get a statistical summary of your data for each category in a group.
df.groupby('Name').describe()
```

#### First step in data cleaning we remove or drop duplicate.

```python
# we can see in 19 and 20 row are same
df = df.drop_duplicates() # we create a variable df now this is aur new dataframe
df
```

#### Now we remove or drop the column that we don't need like ''Not_Useful_Column''

```python
# Here we can see columns = 'Not_Useful_Column' is removed from dataframe
df = df.drop(columns = 'Not_Useful_Column')
df
```

#### Now we remove extra symbol in Last_Name value or anyother value.

```python

df['Last_Name'].str.strip()
# here in row no 2, 11, 14 some extra charector
```

### The lstrip() method is used to remove leading **(left-side)** characters from a string

```python
## The lstrip() method is used to remove leading (left-side) characters from a string
df['Last_Name'].str.lstrip('...') # remove ...potter to potter
```

### The rstrip() method is used to remove leading **(right-side)** characters from a string

```python
# In Row 14    Flenderson remove _ .
df['Last_Name'].str.rstrip('_')
```

### Now remove all extra data char

```python
# df['Last_Name'] is a column we use df['Last_Name']=df['Last_Name'].str.lstrip('...') to change in perticuler column

df['Last_Name']=df['Last_Name'].str.lstrip('...')
df['Last_Name']=df['Last_Name'].str.lstrip('/')
df['Last_Name']=df['Last_Name'].str.rstrip('_')

#print all dataframe
df 
```

```python
df["Last_Name"] = df["Last_Name"].str.strip('123._/')# we can also use this
df
```

### Now its time to clean phone number data correct all number in same type or formate

âœ… Why use regex=True?
Because you're using a regular expression pattern like:

python

'[^a-zA-Z0-9]'
This pattern means:

[^...] â†’ Negation: match anything not in the brackets.

a-zA-Z0-9 â†’ all letters and digits.

So it finds all characters except letters and numbers â€” like @, -, +, (, ), etc.

To use that kind of pattern, pandas needs to know it's a regex pattern, so we pass:*

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '',regex=True)
df
```
